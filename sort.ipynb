СОРТИРОВКА ПУЗЫРЬКОМ

```python
w = 0 #переменная для количества иттераций при сортировке пузырьком
from random import randint#импортируем библиотеку
list_1=[]#создаем пустой список
for i in range (0, 100): #цикл для добавления элементов в список
    a = randint(1, 1_000_000) #выбираем число
    list_1.append(a) #добавляем его в список
stop_flag=True#нужен для того чтобы остановить цикл когда нужно
while (stop_flag is True):
    stop_flag=False#сразу же выкл его
    for i in range (0,len(list_1)-1):#проходим попарно по всему списку
        if list_1[i]<list_1[i+1]:
            a=list_1[i]#вводим переменную для того чтобы могли туда перекидывать значения
            list_1[i]=list_1[i+1]
            list_1[i+1]=a
            stop_flag=True #включаем чтобы он смог еще раз зайти в цикл
            w += 1 
    print(f'{w}. {list_1}')
    
```

ВЫБРОЧНАЯ СОРТИРОВКА

```python
list_2=[]#создаем пустой список
q = 0#переменная для количества иттераций при выборочной сортировке
for i in range (0, 100): #цикл для добавления элементов в список
    b = randint(1, 1_000_000) #выбираем число
    list_2.append(b) #добавляем его в список
sorted_list=[]#создаем пустой список,где будут уже отсортированные значения
while len(list_2)>0:#цикл будет работать, пока в изначальном списке будут элементы
    min=list_2[0]
    for i in range(0,len(list_2)):
        if list_2[i]>min:
            min=list_2[i]
    sorted_list.append(min)
    list_2.remove(min)
    q += 1
    print(f'{q}. {sorted_list}')
```

```python
print(f'Для сортировки списка пузырьком нам понадобилось {w-1} иттераций\nДля сортирочки списка выборкой нам понадобилось {q} иттераций')
#пишем w-1 так при сортировке пузырьком поселдний цикл остается таким же как и предыдущий
```
